# RAG

### Loading

```python
import os
```

```python
#this must be done before importing WebBaseLoader
os.environ["USER_AGENT"] = "Mozilla/5.0 (Machintosh; Intel Mac OS X 10_15_7) AppleWebKit/737.36 (KHTML, , like Gecko) Chrome/91.0.4472.124 Safari/537.36"
```

```python
from langchain_community.document_loaders import WebBaseLoader
```

```python
#대한민국 국가대표
url = "https://www.kfa.or.kr/national/?act=nt_man"
```

```python
import bs4
```

```python
#using BeautifulSoup
loader = WebBaseLoader(
    web_path = url,
    verify_ssl = False,
    bs_kwargs = dict(
        parse_only = bs4.SoupStrainer(
            class_= ("left_dataTable")
        )
    ),
)
```

```python
page = loader.load()
```

```python
page
```

### Splitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 100
)
```

```python
texts = text_splitter.split_documents(page)
```

```python
texts
```

```python
len(texts)
```

### Embedding

```python
from langchain_openai import OpenAIEmbeddings
```

```python
embedding_model = OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
from langchain.vectorstores import Chroma
```

```python
### 내용저장하기
db = Chroma.from_documents(
    documents = texts,
    embedding = embedding_model,
    persist_directory = "./chroma_db"
)
```

```python
### 내용 불러오기
db = Chroma(
    persist_directory = "./chroma_db",
    embedding_function = embedding_model
)
```

### Maximal Marginal Relevance

```python
#MMR(Maxomal Marginal Relevance)
retriever = db.as_retriever(
    search_type = "mmr",
    search_kwargs = {"k":30, "fetch_k": 30, "lambda_mult": 0.8}
)
```

```python
#Default
question = "대한민국 국가대표 축구선수 26명 모두 이름 순서대로 소속팀과 함께 알려줘."
```

```python
context = retriever.invoke(question)
```

```python
context
```

### Augment

```python
from langchain.prompts import ChatPromptTemplate
from langchain.prompts import HumanMessagePromptTemplate
from langchain_core.messages import SystemMessage
```

```python
chat_template = ChatPromptTemplate.from_messages(
    [
        SystemMessage(
            content = "당신은 대한민국 국가대표 기자입니다."
        ),
        HumanMessagePromptTemplate.from_template(
            """
            {question}
            아래의 문맥에 기반하여 답해주세요.
            {content}
            """
        )
    ]
)
```

```python
message = chat_template.format_messages(
    question = question,
    content = context
)
```

```python
from langchain_openai import ChatOpenAI
```

```python
model = ChatOpenAI(
    model_name = "gpt-4o-mini",
    temperature = 0
)
```

```python
for chunk in model.stream(message) :
    print(chunk.content, end = "", flush = True)
```
